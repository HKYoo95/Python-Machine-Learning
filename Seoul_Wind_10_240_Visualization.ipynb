__1시간__

```python
# 서울 10m 데이터 전체 로드하고 datetime으로 index 설정
temp = seoul10.set_index('datetime',inplace=False)

# 격자당 5월 30일 12시부터 21시 까지 10시간 데이터만 추출
temp2 = temp['2020-05-30 12:00:00']

# GeoJson
wind10_1h_gpd = gpd.GeoDataFrame(temp2.reset_index().drop('geometry',axis=1),\
                                 geometry=gpd.points_from_xy(x=temp2.longitude,y=temp2.latitude))

# datetime이 유닉스시간으로 저장되는 것을 방지하기 위해 datetime 컬럼 String으로 변경
wind10_1h_gpd['datetime'] = wind10_1h_gpd['datetime'].astype('str')

# geojson 저장
wind10_1h_gpd.to_file('wind10_1h.geojson',driver='GeoJSON')

# JSON
# temp2.to_json('seoul_wind10m_1h.json',orient='records')
```

__3시간__

```python
# 서울 10m 데이터 전체 로드하고 datetime으로 index 설정
temp = seoul10.set_index('datetime',inplace=False)

# 격자당 5월 30일 12시부터 21시 까지 10시간 데이터만 추출
temp2 = temp['2020-05-30 12:00:00':'2020-05-30 14:00:00']

# 날짜시간으로 정렬하기 위해 인덱스 reset
temp2.reset_index(inplace=True)

# 날짜시간으로 정렬하고 뒤죽박죽 섞인 인덱스를 없애기 위해 다시 날짜시간으로 인덱스 설정 후 
# geojson
temp2 = temp2.sort_values('datetime').drop('geometry',axis=1).set_index('datetime').reset_index()

# JSON
# 0부터 인덱스를 넣기 위해 두번 reset 인덱스.
#temp2 = temp2.sort_values('datetime').drop('geometry',axis=1).set_index('datetime').reset_index().reset_index()

# json으로 저장시 datetime이 유닉스시간으로 저장되는 것을 방지하기 위해 datetime 컬럼 String으로 변경
temp2['datetime'] = temp2['datetime'].astype('str')

# csv 저장
#temp2.to_csv('wind10_3h.csv',index=False)

# 행기준 json 저장
#temp2.to_json('seoul_wind10m_3h.json',orient='records')
wind10_3h_gpd = gpd.GeoDataFrame(temp2,geometry=gpd.points_from_xy(x=temp2.longitude,y=temp2.latitude))
wind10_3h_gpd.to_file('wind10_3h.geojson',driver='GeoJSON')
```
